2. Страховая компания провела исследование с участием 1000 человек и обнаружила, что:
    
    a) 706 человек выбрали как минимум полис страхования жизни.
    
    b) 445 человек выбрали как минимум полисы страхования жизни и автомобиля.
    
    c) 75 человек выбрали как минимум полис страхования жилья.
    
    d) 913 человек выбрали как минимум один вид страхового полиса.
    
    Мы не можем с уверенностью предсказать выбор людей. Поэтому вы можете предположить, что выбор человека для покупки страхового полиса является случайным экспериментом. В исследовании приняли участие 1000 человек, поэтому можно считать, что случайный эксперимент был выполнен 1000 раз. Найдите:
    
    * вероятность того, что человек выберет хотя бы страхование жизни. Присвойте ответ переменной ```p_life```.
    
    * вероятность того, что человек выберет хотя бы полисы страхования жизни и автомобиля. Присвойте ответ переменной ```p_life_car```.
    
    * вероятность того, что человек выберет хотя бы полис страхования жилья. Присвойте ответ переменной ```p_house```.
    
    * вероятность того, что человек вообще не будет страховаться. Присвойте ответ переменной ```p_no```.

````python
#Начало вашего кода
total = 1000
at_least_life = 706
at_least_life_car = 445
at_least_house = 75
at_least_one_insurance = 913
p_life = at_least_life / total
p_life_car = at_least_life_car / total
p_house = at_least_house / total

p_no = 1 - (at_least_one_insurance / total)
#Конец вашего кода

print(f"{p_life*100:.2f} %")
print(f"{p_life_car*100:.2f} %")
print(f"{p_house*100:.2f} %")
print(f"{p_no*100:.2f} %")
````

3. Пожалуйста, смоделируйте эксперименты с монетой. Предположим, что есть одинаковый шанс получить Heads и Tails после одного броска монеты. Оцените вероятность получения Tails, если вы подбрасываете монету от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:




````python
#Начало вашего кода
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

num_flips = np.arange(10, 10001)
p_tails = np.zeros_like(num_flips, dtype=float) 
for i, flips in enumerate(num_flips):
    outcomes = np.random.choice(['Heads', 'Tails'], size=flips)  
    num_tails = np.sum(outcomes == 'Tails')
    p_tails[i] = num_tails / flips  
    
plt.figure(figsize=(6, 4))
plt.plot(num_flips, p_tails, label='Simulation')
plt.axhline(y=0.5, color='r', linestyle='--', label='Theorectical')
plt.xlabel('# of tails')
plt.ylabel('Probability Tails')
plt.title('P(Tails)')
plt.legend()
plt.xlim(left=10)
plt.show()
#Конец вашего кода
````

4. Пожалуйста, смоделируйте эксперименты с игральной костью (со сторонами ```{1,2,3,4,5,6}```). Предположим, что есть равный шанс получить любую из шести сторон после броска игральной кости. Оцените вероятность получить число меньше или равное 2 после броска игральной кости от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:



````python
#Начало вашего кода
np.random.seed(1)  # Установка seed для воспроизводимости

num_rolls = np.arange(10, 10001)
p_less_equal_2 = np.zeros_like(num_rolls, dtype=float)

for i, rolls in enumerate(num_rolls):
    outcomes = np.random.randint(1, 7, size=rolls)  
    num_less_equal_2 = np.sum(outcomes <= 2) 
    p_less_equal_2[i] = num_less_equal_2 / rolls 

plt.figure(figsize=(7, 5))
plt.plot(num_rolls, p_less_equal_2, label='Simulation')
plt.axhline(y=2/6, color='r', linestyle='--', label='Theoretical')
plt.xlabel('# of flips')
plt.ylabel('Probability')
plt.title('P(<3)')
plt.legend()
plt.xlim(left=10)
plt.show()
#Конец вашего кода
````

5. Пожалуйста, смоделируйте эксперименты с игральной костью (со сторонами ```{1,2,2,3,4,5}```). Предположим, что есть равный шанс получить любую из шести сторон после броска игральной кости. Оцените вероятность получить число меньше или равное 3 после броска игральной кости от 10 до 10000 раз. Размер фигуры ```(7,5)``` и ```seed=1```

    Вы можете визуализировать результаты моделирования следующим образом:

### Ожидаемые результаты:



````python
#Начало вашего кода
np.random.seed(1)  # Установка seed для воспроизводимости

num_rolls = np.arange(10, 10001)
p_less_equal_3 = np.zeros_like(num_rolls, dtype=float)

for i, rolls in enumerate(num_rolls):
    outcomes = np.random.choice([1, 2, 2, 3, 4, 5], size=rolls)
    num_less_equal_3 = np.sum(outcomes <= 3)  
    p_less_equal_3[i] = num_less_equal_3 / rolls  
plt.figure(figsize=(7, 5))
plt.plot(num_rolls, p_less_equal_3, label='Simulation')
plt.axhline(y=4/6, color='r', linestyle='--', label='Theoretical')
plt.xlabel('# of flips')
plt.ylabel('Probability')
plt.title('P(<4)')
plt.legend()
plt.xlim(left=10)
plt.show()
#Конец вашего кода
````
